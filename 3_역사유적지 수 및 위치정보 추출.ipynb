## 1. 데이터 불러오기

### csv파일 불러오기

```python
df = pd.read_csv('data/3/국내 지역별 유적지.csv')
df
```

## 2. 각 지역별 역사유적지 수 추출 및 저장

- group by를 이용해 개수 확인

```python
# CTPRVN_NM(시도명), SIGNGU_NM(시군구명) 컬럼을 그룹으로 묶어 갯수를 확인
df_count = df.groupby(['CTPRVN_NM','SIGNGU_NM']).size().reset_index(name='고궁/역사 유적지 수')
df_count
```

- 컬럼명 변경

```python
df_count.rename(columns={'CTPRVN_NM': '시도명','SIGNGU_NM':'시군구명'}, inplace=True)
df_count
```

```python
# null값 확인
df_count.isnull().sum()
```

- csv파일 저장

```python
df_count.to_csv('./data/3/전처리_완료/지역별 역사유적지지 수.csv', sep=',', na_rep='NaN')
```

----------

## 3. 각 지역별 역사유적지 정보(상호명 및 경/위도) 추출 및 저장

- 필요한 컬럼('POI_NM', 'CTPRVN_NM','SIGNGU_NM', 'LC_LO', 'LC_LA') 추출

```python
df_loc = df[['POI_NM', 'CTPRVN_NM','SIGNGU_NM', 'LC_LO', 'LC_LA']]
df_loc
```

- 인덱스명 변경

```python
df_loc.rename(columns={'POI_NM': '유적지명','CTPRVN_NM': '시도명','SIGNGU_NM':'시군구명','LC_LO':'경도','LC_LA':'위도'}, inplace=True)
df_loc
```

- csv파일로 저장

```python
# csv파일로 저장
df_loc.to_csv('./data/3/전처리_완료/지역별 역사유적지 위치정보.csv', sep=',', na_rep='NaN')
```
