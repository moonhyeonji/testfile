```python
# 여러 페이지에 있는 데이터를 하나의 리스트에 모으기
itemlist=[]
for page in tqdm(pagelist):
    items=page.find_all('item')
    for item in items:
        itemlist.append(item)
```

```python
itemlist[0]
```

```python
# 필요한 컬럼명
columns=["signguCode","signguNm","touDivNm","touNum",'baseYmd']
```

```python
# openAPI로 받은 데이터에서 필요한 컬럼만 추출하여 새로운 리스트에 저장하는 함수
def new_function(item_tag):
    columns=["signguCode","signguNm","touDivNm","touNum",'baseYmd']
    estate={}
    for col in columns:
        estate[col]=item_tag.find(col).text
    return estate
```

```python
# 필요한 컬럼만 추출하여 새로운 리스트에 저장
estate_list=[]
for item in tqdm(itemlist):
    estate=new_function(item)
    estate_list.append(estate)
estate_list
```

```python
# 리스트에 있는 데이터를 새로운 데이터프레임을 생성
df=pd.DataFrame(estate_list, columns=columns)
df.head(30)
```

```python
# 완성한 데이터프레임 저장하기
df.to_csv('관광방문객전체.csv')
```
