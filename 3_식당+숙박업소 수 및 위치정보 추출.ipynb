### 식당 데이터 추출

```python
# 상권업종대분류명이 '음식'인 데이터만 분류하기
df_restaurant = df[df.상권업종대분류명 == '음식']
```

### 숙박업소 데이터 추출

```python
# 인덱스 재설정
df_accommodation.reset_index(drop=True, inplace=True)
```

## 4. 데이터 저장

```python
df_restaurant_group = df_restaurant.groupby(['시도명', '시군구명'])[['상가업소번호']].count().reset_index()
df_restaurant_group
```

```python
df_restaurant_group.rename(columns={'상가업소번호': '식당 수'}, inplace=True)
df_restaurant_group
```

```python
df_restaurant_group.to_csv('./data/3/전처리_완료/지역별 식당 수.csv', sep=',', na_rep='NaN')
```

```python
df_accommodation_group = df_accommodation.groupby(['시도명', '시군구명'])[['상가업소번호']].count().reset_index()
df_accommodation_group
```

```python
df_accommodation_group.rename(columns={'상가업소번호': '숙박업소 수'}, inplace=True)
df_accommodation_group
```

```python
df_accommodation_group.to_csv('./data/3/전처리_완료/지역별 숙박업소 수.csv', sep=',', na_rep='NaN')
```
