## 1. 교통수단

### 데이터 불러오기 

```python
df = pd.read_csv('./data/4-1/4-1_외국인주요교통수단_20172019.csv', encoding='cp949')
df
```

- 데이터 확인하기 

```python
# 정보확인
df.info()
```

```python
# null값 확인
df.isnull().sum()
```

### 필요한 데이터만 추출

 - 불필요한 행 삭제

```python
# 필요한 행만 추출 (1번행)
df = df.loc[[0,1]]
df
```

 - 불필요한 열 삭제

```python
# 필요한 열 추출 (월별 컬럼 삭제, 특성컬럼 삭제)
df_2017 = df[['2017.1','2017.2','2017.3','2017.4','2017.5','2017.6','2017.7',
             '2017.8']]
df_2017
```

### 데이터 정리

- Trans하여 보기 쉽게 DataFrame정리

```python
# 행과열 뒤집기
df_2017=df_2017.T
df_2017
```

- 인덱스변경

```python
# 교통수단을 인덱스로 변경
df_2017 = df_2017.set_index(0)
df_2017
```

- 컬럼명 변경

```python
# 컬럼이름 변경
df_2017 = df_2017.rename({1:'2017년'}, axis='columns')
df_2017
```

- 각 년도마다 하나의 행으로 정리 (2017년, 2018년, 2019년)

```python
# 하나의 행으로 2017년 데이터 정리
df_2017 = df_2017.T
df_2017 = df_2017.rename({1:'2017년'}, axis='index')
df_2017
```

```python
# 위과정 반복하여 2018데이터 생성
df_2018 = df[['2018.1','2018.2','2018.3','2018.4','2018.5','2018.6','2018.7',
             '2018.8']]

df_2018=df_2018.T

df_2018 = df_2018.set_index(0)

df_2018 = df_2018.T
df_2018=df_2018.rename({1:'2018년'}, axis='index')
```

```python
# 위과정 반복하여 2019데이터 생성
df_2019 = df[['2019.1','2019.2','2019.3','2019.4','2019.5','2019.6','2019.7',
             '2019.8']]

df_2019=df_2019.T

df_2019 = df_2019.set_index(0)

df_2019 = df_2019.T
df_2019=df_2019.rename({1:'2019년'}, axis='index')
```

- 2017, 2018, 2019년도 행 합치기

```python
# 2017년, 2018년, 2019년 데이터 합치기
data = pd.concat([df_2017,df_2018,df_2019])
data
```

```python
# 행열 전환
data=data.T
```

- 인덱스 명 변경

```python
# 인덱스명 변경
data.index.name = '교통수단'
data
```

### 그래프 생성
- 택시 이용률이 눈에 띄게 많다는 것을 확인

```python
data = data.astype(float)
data.plot(kind='bar',color=['tan','navajowhite','silver'])
```
